```python
import pandas as pd

df = pd.read_csv('/content/insurance.csv')
df.head()
```

```python
import numpy as np

# Mengambil fitur BMI dan target charges
X = df['bmi'].values
y = df['charges'].values

# Normalisasi data
X = (X - X.mean()) / X.std()

# Inisialisasi parameter
m = 0
c = 0

learning_rate = 0.01
epochs = 1000
n = len(X)

# Proses SGD manual
for epoch in range(epochs):
    for i in range(n):
        y_pred = m * X[i] + c

        # Gradient
        dm = -2 * X[i] * (y[i] - y_pred)
        dc = -2 * (y[i] - y_pred)

        # Update parameter
        m = m - learning_rate * dm
        c = c - learning_rate * dc

print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_pred_all = m * X + c

sorted_index = np.argsort(X)
X_sorted = X[sorted_index]
y_pred_sorted = y_pred_all[sorted_index]

plt.figure(figsize=(10, 6))
plt.scatter(X, y, alpha=0.6, s=50, marker='o', label='Data Asli')
plt.plot(X_sorted, y_pred_sorted, linewidth=3, label='Garis Regresi SGD')

plt.title('Visualisasi Regresi Linear Menggunakan SGD', fontsize=14)
plt.xlabel('BMI (sudah dinormalisasi)', fontsize=12)
plt.ylabel('Charges', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```
